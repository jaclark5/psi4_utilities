```python
import psi4 

qce_mol = psi4.core.Molecule.from_string("""
0 3
H  0.0  0.0  0.0
H  0.0  0.0  0.74
O  0.0  0.74  0.0
""")
```

```python
import psi4_memory as pm

pm.memory_from_psi4(qce_mol, method="b2plyp-d3", basis="dzvp", use_wfn=True)
```

```python
psi4_mol = psi4.geometry(qce_mol.to_string(dtype='psi4'))
# Set options only if values are not None
options = {
    key: value for key, value in {
        'basis': "aug-cc-pVTZ-dk",
        'print': 5,
        'reference': 'uks',
    }.items() if value is not None
}
psi4.set_options(options)
```

```python
e, wfn = psi4.energy('scf', return_wfn=True)
```

```python
psi4.oeprop(wfn, 'DIPOLE', 'LOWDIN_CHARGES', title='H3O+ SCF')
```
