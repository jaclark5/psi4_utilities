# Test: Basis Set Exchange Utilities

This notebook tests the `bse_utilities` module, which provides tools for querying and filtering basis sets from the Basis Set Exchange (BSE).

Test coverage:
1. Finding auxiliary basis sets by orbital basis prefix
2. Retrieving basis sets supporting specific elements
3. Filtering by ECP (effective core potential) availability
4. Verbose output with basis set details

## Setup: Import Required Libraries

```python
# Import required modules
from psi4_utilities.bse_utilities import find_aux_sets, get_basis_sets
```

## Test 1: Find Auxiliary Basis Sets
Search for auxiliary basis sets (jkfit, mp2fit, ri, cabs) corresponding to orbital basis prefixes.

```python
# Test finding auxiliary basis sets for cc-pVTZ
aux_sets_ccpvtz = find_aux_sets("cc-pVTZ")
print(f"Auxiliary basis sets for cc-pVTZ:")
for basis in sorted(aux_sets_ccpvtz):
    print(f"  - {basis}")
```

```python
# Test finding auxiliary basis sets for another orbital basis
aux_sets_sto = find_aux_sets("sto")
print(f"\nAuxiliary basis sets for sto:")
for basis in sorted(aux_sets_sto):
    print(f"  - {basis}")
```

## Test 2: Get Basis Sets by Elements
Retrieve basis sets supporting specific elements.

```python
# Test with element symbols
basis_sets_ch = get_basis_sets(elements=["C", "H"])
print(f"Basis sets supporting C and H (first 10):")
for basis in sorted(basis_sets_ch)[:10]:
    print(f"  - {basis}")
print(f"\nTotal: {len(basis_sets_ch)} basis sets")
```

```python
# Test with atomic numbers
basis_sets_ch_atomicnum = get_basis_sets(elements=[6, 1])
print(f"Basis sets supporting elements 6 (C) and 1 (H) match above: {basis_sets_ch == basis_sets_ch_atomicnum}")
```

## Test 3: Filter by ECP Availability
Test the `has_ecp` parameter to filter basis sets by ECP presence.

```python
# Get all basis sets for a transition metal
all_I = get_basis_sets(elements=["I"])
print(f"Total basis sets for I: {len(all_I)}")

# Get only basis sets WITH ECPs
I_with_ecp = get_basis_sets(elements=["I"], has_ecp=True)
print(f"Basis sets for I WITH ECPs: {len(I_with_ecp)}")
for basis in sorted(I_with_ecp)[:5]:
    print(f"  - {basis}")

# Get only basis sets WITHOUT ECPs
I_no_ecp = get_basis_sets(elements=["I"], has_ecp=False)
print(f"Basis sets for I WITHOUT ECPs: {len(I_no_ecp)}")
for basis in sorted(I_no_ecp)[:5]:
    print(f"  - {basis}")
```

## Test 4: Verbose Mode
Display detailed information about basis sets including descriptions, basis function counts, and ECP details.

```python
# Test verbose mode with multiple elements to show ECP header
print("Basis sets for Fe and Ni WITH ECPs (verbose - first 3):")
results = get_basis_sets(elements=["Fe", "Ni"], has_ecp=True, verbose=True)
```

```python
# Test verbose mode with ECP filtering for transition metal
print("Basis sets for Pd WITH ECPs (verbose - first 5):")
results = get_basis_sets(elements=["Pd"], has_ecp=True, verbose=True)
print(f"\nTotal: {len(results)} basis sets with ECPs for Pd")
```

```python
# First, check what's available for Pd overall
all_pd = get_basis_sets(elements=["Pd"])
pd_with_ecp = get_basis_sets(elements=["Pd"], has_ecp=True)
pd_no_ecp = get_basis_sets(elements=["Pd"], has_ecp=False)

print(f"Total basis sets for Pd: {len(all_pd)}")
print(f"Basis sets for Pd WITH ECPs: {len(pd_with_ecp)}")
print(f"Basis sets for Pd WITHOUT ECPs: {len(pd_no_ecp)}")
if pd_with_ecp:
    print(f"Pd ECP basis sets: {pd_with_ecp[:5]}")
```

## Test 5: Verify ECP Filtering Logic
Confirm that filtering by ECP status works correctly.

```python
# Verify that has_ecp=True and has_ecp=False don't overlap
all_basis = set(get_basis_sets(elements=["Fe"], has_ecp=None))
with_ecp = set(get_basis_sets(elements=["Fe"], has_ecp=True))
no_ecp = set(get_basis_sets(elements=["Fe"], has_ecp=False))

print(f"All basis sets: {len(all_basis)}")
print(f"With ECP: {len(with_ecp)}")
print(f"Without ECP: {len(no_ecp)}")
print(f"With + Without = All: {len(with_ecp) + len(no_ecp) == len(all_basis)}")
print(f"No overlap: {len(with_ecp & no_ecp) == 0}")
```
